# Data Preprocessing

## Import Library

```python
import numpy as np
import pandas as pd
```

## Import Data 

```python
pd.set_option('display.max_columns',500)
df = pd.read_excel("Indeks_pencemaran_2021.xlsx")
df
```

## Cek Duplicate 

```python
df['tanggal'].duplicated().sum()
```

## Cek dan Hapus Null Value 

```python
df.isnull().sum()
```

```python
df['pm25'] = df['pm25'].fillna(df['pm25'].mean())
df.tail(35)
```

## Cek dan Ubah Tipe Data  

```python
df.dtypes
```

```python
df['tanggal']= df['tanggal'].apply(pd.to_datetime)
```

```python
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
df['categori'] = Le.fit_transform(df['categori'])
df
```

```python
df['pm25'] = df['pm25'].astype(np.int64)
```

```python
df.dtypes
```

## Memilih Data yang Digunakan 

## 1.Untuk Naive Bayes 

```python
df_naive = df[['pm10','pm25','so2','co','o3','no2','categori']]
df_naive
```

## 2.Untuk Decision Tree 

```python
df_dtree = df[['pm10','pm25','so2','co','o3','no2','categori']]
df_dtree
```

## 3.Untuk KNeighbors 

```python
df_kn = df[['pm10','pm25','so2','co','o3','no2','categori']]
df_kn
```

# Standard Naive Bayes

## Import Naive Bayes 

```python
from sklearn.model_selection import train_test_split

features = df_naive.iloc[:,:6]
label = df_naive.iloc[:,6]

x_train, x_test, y_train, y_test = train_test_split(features,label,test_size=0.30, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model = model.fit(x_train, y_train)
```

## Prediksi Naive Bayes 

```python
print(model.predict(x_test))
print(y_test)
```

## Confusion Matrix Naive Bayes 

```python
from sklearn.metrics import confusion_matrix
hasil_prediksi = model.predict(x_test)
confusion_matrix(hasil_prediksi, y_test)
```

## classification Report  Naive Bayes

```python
from sklearn.metrics import classification_report
print(classification_report(hasil_prediksi, y_test))
```

# Standard Decision Tree

## Import Decision Tree 

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
```

```python
atr_data = df_dtree.drop(columns = 'categori')
atr_data.head()
```

```python
cls_data = df_dtree[ 'categori']
cls_data.head()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(atr_data, cls_data, test_size=0.3, random_state=0)
tree_data = DecisionTreeClassifier(random_state=0)
tree_data.fit(xtrain, ytrain)
```

## Prediksi Decision Tree 

```python
print(tree_data.predict(xtest))
print(ytest)
```

## Confusion Matrix Decision Tree

```python
from sklearn.metrics import confusion_matrix
prediksi = tree_data.predict(xtest)
confusion_matrix(prediksi, ytest)
```

## classification Report Decision Tree

```python
from sklearn.metrics import classification_report
print(classification_report(prediksi, ytest))
```

# Standard KNeighbors

## Import KNeighbors

```python
from sklearn.model_selection import train_test_split

X = df_kn.iloc[:,:6].values
Y = df_kn.iloc[:,6].values

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.30, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2, metric = 'euclidean')
knn.fit(X_train,Y_train)
pred_i = knn.predict(X_test)
```

## Prediksi KNeighbors

```python
print(pred_i)
print(Y_test)
```

## Confusion Matrix KNeighbors

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(pred_i, Y_test)
```

## classification Report KNeighbors

```python
from sklearn.metrics import classification_report
print(classification_report(pred_i, Y_test))
```

# Perbandingan Akurasi

```python
from sklearn.metrics import accuracy_score
print("Akurasi Naive Bayes : ",accuracy_score(y_test, hasil_prediksi))
print("Akurasi Decision Tree : ",accuracy_score(ytest, prediksi))
print("Akurasi KNeighbors : ",accuracy_score(Y_test, pred_i))
```

# Data Modeling

```python
from jcopml.plot import plot_confusion_matrix
print("Matrixs Naive Bayes", plot_confusion_matrix(x_train, y_train, x_test, y_test, model))

print("Matrixs Decision Tree", plot_confusion_matrix(xtrain, ytrain, xtest, ytest, tree_data))

knn_data =  knn.fit(X_train, Y_train)
print("Matrixs KNeighbors",plot_confusion_matrix(X_train, Y_train, X_test, Y_test, knn_data))
```
